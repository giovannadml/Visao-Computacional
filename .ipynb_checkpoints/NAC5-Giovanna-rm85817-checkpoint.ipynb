### **Segmentando uma imagem**

Segmentar apenas os círculos maiores da imagem e exibir o resultado. Exibe na imagem a área e o centro de massa dos dois círculos segmentados, resultando em uma imagem com um marcação no centro indicando o CM e o valor da área.


```python
%matplotlib inline
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np

img = cv.imread('circulo.png', 0)
img_rgb = cv.imread('circulo.png')
cimg = cv.cvtColor(img_rgb,cv.COLOR_RGB2BGR)

circles = cv.HoughCircles(img,cv.HOUGH_GRADIENT,1,20,
                            param1=50, param2=30, minRadius=0, maxRadius=110)

circles = np.uint16(np.around(circles))
for i in circles[0,:]:
    cv.circle(cimg,(i[0],i[1]),2,(255, 255, 255), 250)


plt.figure(figsize = (15, 15))
plt.imshow(cimg)
plt.show()
```

```python
%matplotlib inline
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np

img = cv.imread('circulo.png', 0)
img_rgb = cv.imread('circulo.png')
cimg = cv.cvtColor(img_rgb,cv.COLOR_RGB2BGR)

circles = cv.HoughCircles(img,cv.HOUGH_GRADIENT,1,20,
                            param1=50, param2=30, minRadius=0, maxRadius=110)

circles = np.uint16(np.around(circles))
for i in circles[0,:]:
    cv.circle(cimg,(i[0],i[1]),2,(255, 255, 255), 250)

gray = cv.cvtColor(cimg, cv.COLOR_BGR2GRAY)
blurred = cv.GaussianBlur(gray, (5, 5), 0)
thresh = cv.threshold(blurred, 60, 255, cv.THRESH_BINARY)[1]

contornos = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)[0]

maior = []
for c in contornos:
    area = cv.contourArea(c)
    maior.append(c)
    print("Área:",area)

    
contornos_img = blurred.copy()
#cv.drawContours(contornos_img, [maior], -1, [20, 130, 90], 15)

cnt = contornos[0]

M = cv2.moments(cnt)


plt.figure(figsize = (15, 15))
plt.imshow(cimg)
plt.show()
```
