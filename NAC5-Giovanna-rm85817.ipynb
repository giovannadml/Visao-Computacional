### **Segmentando uma imagem**

```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np

img = cv2.imread('circulo.png')
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        
plt.figure(figsize = (10, 10))
plt.imshow(image)
plt.show()
```
